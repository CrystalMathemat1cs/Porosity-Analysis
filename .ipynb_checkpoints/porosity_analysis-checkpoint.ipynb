# <div align='center'>Determining the Porosity of Yttria-Stabilised-Zirconia Thermal Barrier Coatings (YSZ-TBCs) Through Image Processing</div>

This notebook applies several image processing techniques to a list of Scanning Electron Microscopy (SEM) images of different samples of TBC cross-sections in order to determine the porosity of the YSZ layer. First the appropriate libraries used are imported below:

```python
import matplotlib.pyplot as plt
from matplotlib import patches
import numpy as np
from skimage import io
from skimage.filters import threshold_otsu, sobel
from skimage.segmentation import flood, flood_fill
```

The images in the 'SEM_images' folder are samples of cross sections of the TBC layers which have been deposited by different methods. We first import the images into a list:

```python
images = []
for i in range(5):   
    image = io.imread("SEM_images\Sample{:d}.jpg".format(i + 1))
    images.append(image)
```

Let's look at an example image:

```python
fig1, ax1 = plt.subplots(1, figsize=(16,9))
ax1.imshow(images[0], cmap=plt.cm.gray)
rect = patches.Rectangle((5,1540), 2030, 230, linewidth=2, edgecolor='r', facecolor='none')
arrow = patches.Arrow(750, 1400, 120, 120, linewidth=1, edgecolor='r')
ax1.add_patch(rect)
ax1.add_patch(arrow)
ax1.set_title('Sample1', fontsize=14)
plt.text(460, 1400, 'TGO layer', fontsize=14);
ax1.axis('off');
```

We want to apply thresholding to this image so that we can differentiate between the black spots in the layer (porosity) and the rest of the image. As we can see, in order to extract any porosity information on the YSZ layer, we have to do some preprocessing on the images. Firstly, the information at the bottom of the image showing the microscope parameters and magnification will be cropped. Secondly, the highlighted Thermally Grown Oxide (TGO) layer as well as the black background at the top of the image need to be masked so that the pixels in those areas do not contribute to the final pixel count after thresholding.

## Preprocessing stage
The function `process(images)` below carries out the preprocessing needed before we can apply masking and finally threshold the YSZ layer. First, it crops out the microscope information. Next, it uses `flood_fill()` in order to fill the unwanted background and the TGO layer with white pixels. In order to facilitate the `flood_fill()` a row of black pixels was added to the top of the image so that the whole black region is one connected region. The same was done below by adding a row of gray pixels with intensity similar to that of the TGO. Flood filling the TGO layer was not quite as trivial as the black background, as there is a higher variation in pixel intensities in this region, therefore it was difficult to determine an optimal tolerance that would turn the whole TGO to white without infiltrating the YSZ layer.

```python
def process(images):
    row_of_black = np.zeros(images[0].shape[1])
    row_of_gray = np.full(shape=images[0].shape[1], fill_value=170, dtype=np.int)
    processed = []
    for image in images:
        image = image[:1760, :]
        image = np.vstack([row_of_black, image])
        image = np.vstack([image, row_of_gray])
        image = flood_fill(image, (0,0), 255, tolerance=30)
        image = flood_fill(image, (image.shape[0]-1, 0), 255, tolerance=40)

        processed.append(image)
    
    return processed
```

```python
processed = process(images)
```

Let's look at Sample1 after preprocessing:

```python
fig2, ax2 = plt.subplots(1, 2, figsize=(16,9))

ax2[0].imshow(images[0], cmap=plt.cm.gray)
ax2[0].set_title('Original Sample1 image', fontsize=14)

ax2[1].imshow(processed[0], cmap=plt.cm.gray)
ax2[1].set_title('Sample1 after preprocessing', fontsize=14)

ax2[0].axis('off');
ax2[1].axis('off');
```

## Masking unwanted regions
Now that we have highlighted the regions we don't want, we can mask them using the `flood()` function. The `find_ysz_layer()` function below implements this on all of the processed images list. I found that applying an edge detection Sobel filter to the image and then applying `flood()` gave better results, as both unwanted regions (background and TGO) get masked this way by only flooding the image once.

```python
def find_ysz_layer(processed_images):
    seed = (0,0)
    ysz_layers = []
    
    for image in processed_images:
        img_sobel = sobel(image)
        mask = flood(img_sobel, seed, tolerance = 0.01)
        ysz_layers.append(mask)
    
    return ysz_layers
```

```python
ysz_layers = find_ysz_layer(processed)
```

The result of the masking is shown below. The unwanted regions are masked and appear white. This new image is now an array in which the white pixels are `True` while the black pixels are `False`. We will use this to apply thresholding only to the unmasked region (the YSZ layer) in the next section.

```python
fig3, ax3 = plt.subplots(1, 3, figsize=(16,9))

ax3[0].imshow(processed[0], cmap=plt.cm.gray)
ax3[0].set_title('Sample1 after processing', fontsize=14)

ax3[1].imshow(sobel(processed[0]), cmap=plt.cm.gray)
ax3[1].set_title('Sobel filter applied', fontsize=14)

ax3[2].imshow(ysz_layers[0], cmap=plt.cm.gray)
ax3[2].set_title('Masked regions highlighted in white', fontsize=14)

ax3[0].axis('off');
ax3[1].axis('off');
ax3[2].axis('off');
```

## Applying thresholding
As mentioned before, we now have an array of masked images, `ysz_layers`, where all the pixels in the unwanted regions (white in the figure above) are labeled as `True`, while all the pixels corresponding to the YSZ layer are labeled as `False`. We will use this fact in order to apply thresholding only to the pixels labeled as `False` by the masking, which correspond to the pixels of the YSZ layer in the images in the `processed` list. Otsu's method for thresholding is used here to find the optimal threshold for classifying the pixels in each image.

We first iterate through both the `processed` and `ysz_layers` lists simultaneously. Then we flatten every image to a one-dimensional array and iterate through the pixels, again simultaneously. We check that the current masked pixel is labeled as `False`. If so, we apply thresholding to the current pixel in the processed image. We go through all the pixels and images and the output will be a list containing lists with the thresholded pixel values, `thresh_images`. These lists are not neccessarily 'images' anymore, however they contain the neccesary information in order to compute the porosity by looking at the ratio of `False` pixels with the total number of pixels.

```python
def apply_threshold(images, ysz_layers):
    thresh_images = []
    
    for image, layer in zip(images, ysz_layers):
        threshold = threshold_otsu(image)
        thresh_image = []
        
        for pixel, masked_pixel in zip(image.flatten(), layer.flatten()):
            if masked_pixel == False:
                pixel = pixel > threshold
                thresh_image.append(pixel)
                
        thresh_images.append(thresh_image)
    
    return thresh_images
```

```python
thresh_images = apply_threshold(processed, ysz_layers)
```

Here is an example of a thresholded image using Otsu's method for determining the threshold value, where the pixels are classified as `True` or `False` (white or black) depending on if their intensity is lower or higher than the threshold value (In this case, thresholding is applied to the whole image. When thresholding to determine porosity, the threshold is only applied to the unmasked YSZ layer as explained above):

```python
fig4, ax4 = plt.subplots(1, figsize=(16,9))

ax4.imshow(processed[0] > threshold_otsu(processed[0]), cmap=plt.cm.gray)
ax4.set_title('Sample1 thresholded using Otsu\'s method')
ax4.axis('off');
```

## Determining porosity
Now that we have the lists in `thresh_images`, the last step is to just calculate the ratio of `False` (black) pixels to the total amount of pixels in each list. The function `find_porosity` below takes `thresh_images` and does exactly that for each of our samples.

```python
def find_porosity(layers):
    porosities = []
    for layer in layers:
        porosity = 100 * ((len(layer) - np.sum(layer)) / len(layer))
        porosities.append(round(porosity, 2))
    
    return porosities
```

```python
porosities = find_porosity(thresh_images)
```

```python
for i in range(len(porosities)):
    print('Sample{:d} porosity: '.format(i+1) + str(porosities[i]) + '%')
print('\nAverage porosity: ' + str(round(np.mean(porosities), 2)) + '%')
```
